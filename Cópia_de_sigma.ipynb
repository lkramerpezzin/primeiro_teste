## Cálculo de sigma

```python
def sig(a, b):
  import math
  return math.sqrt((a**2)-(b**2))
```

##Menos Ouro

```python
def difquad(r, o):
  from math import sqrt
  return sqrt( (r**2) - (o**2) )
```

```python
print(sigma_110)
print(Au_sigma_110)
```

```python
explo_110= list()
explo_120= list()
explo_130= list()
```

```python
k = 0
while (k < np.size(Au_sigma_110)):
    if (sigma_110[k] < Au_sigma_110[k] or sigma_120[k] < Au_sigma_120[k] or sigma_130[k] < Au_sigma_130[k]):
        print('o filme está com algum problema em 110° pois os sigmas de H2 é só um pouco maior que H, o que resulte que o sigma do ouro maior, portanto, raiz negativa')
       
    else:
        explo_110.append(difquad(sigma_110[k], Au_sigma_110[k]))
        explo_120.append(difquad(sigma_120[k], Au_sigma_120[k]))
        explo_130.append(difquad(sigma_130[k], Au_sigma_130[k]))

    k += 1
```

```python
j= 0
med110= 0
med120= 0
med130= 0

while(j<np.size(Au_sigma_110) ):
  med110+= Au_sigma_110[j]
  med110= med110/(np.size(Au_sigma_110))
  med120+= Au_sigma_120[j]
  med120= med120/(np.size(Au_sigma_120))
  med130+= Au_sigma_130[j]
  med130= med130/(np.size(Au_sigma_130))
  j+= 1
print(med110, med120, med130)
```

```python
explo_110= list()
explo_120= list()
explo_130= list()

k= 0
while(k<np.size(sigma_110)):
  if k == 1:
    explo_110.append(sigma_110[k])
  else:
    explo_110.append(difquad(sigma_110[k], med110))
  explo_120.append(difquad(sigma_120[k], med120))
  explo_130.append(difquad(sigma_130[k], med130))
  k+= 1


print(med110, med120, med130)
print(explo_110)
print(explo_120)
print(explo_130)
```
